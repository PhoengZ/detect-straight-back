```python
import paho.mqtt.client as mqtt
import numpy as np
import json
import time

# device for testing sending

CLIENT_ID = ""
TOKEN = ""
SECRET = ""
HOST = "mqtt.netpie.io"
PORT = 1883

TOPIC_DATA = "@msg/sensor/distance"
```

```python
def connection(client, userdata, flags, rc):
  if rc == 0:
    print("Successfully connected to netpie")
  else:
    print("Failed to connected to netpie")
```

```python
# simulate distance from test client
client = mqtt.Client(client_id=CLIENT_ID, protocol=mqtt.MQTTv311)
client.username_pw_set(TOKEN, SECRET)
client.on_connect = connection

client.connectแ(HOST,PORT,60 ) # ping to netpie every 60 second to make sure netpie will not disconnect on this device
client.loop_start()

try:
  while True:
    rand_dis = np.random.rand()*20
    payload_dict = {
        "d": rand_dis,
        "b": 1
    }
    payload_str = json.dumps(payload_dict)
    client.publish(TOPIC_DATA, payload_str)
    print(f"already Sent: {payload_str}")
    time.sleep(1)
except KeyboardInterrupt:
  print("\n แStopped by User")
  client.loop_stop()
  client.disconnect()
```
